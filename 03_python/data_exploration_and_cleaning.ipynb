```python
import pandas as pd
import numpy as np
from IPython.display import display
```

# **Load Dataset**

```python
customers           = pd.read_csv(r"olist_customers_dataset.csv")
geolocation         = pd.read_csv(r"olist_geolocation_dataset.csv")
items               = pd.read_csv(r"olist_order_items_dataset.csv")
payment             = pd.read_csv(r"olist_order_payments_dataset.csv")
o_reviews           = pd.read_csv(r"olist_order_reviews_dataset.csv")
orders              = pd.read_csv(r"olist_orders_dataset.csv")
products            = pd.read_csv(r"olist_products_dataset.csv")
product_category    = pd.read_csv(r"product_category_name_translation.csv")
sellers             = pd.read_csv(r"olist_sellers_dataset.csv")
```

# **Automated Exploratory Data Analysis (EDA)**
## **Loop through all datasets to check for quality issues (nulls, types, duplicates)**

```python

dataset = {
    "customers": customers,
    "geolocation": geolocation,
    "orders_items" : items,
    "payment": payment,
    "order_reviews": o_reviews,
    "orders": orders,
    "products": products,
    "product_category": product_category,
    "sellers": sellers
}

for name, df in dataset.items():
    print("\n" + "=" *150)
    print(f'Dataset: {name}')
    print("=" *150)

    # Display dataset dimensions
    print("\n [1] Shape")
    shape = pd.DataFrame(
        {"N_Rows": [df.shape[0]], "N_columns":[df.shape[1]]}
    )
    display(shape)

    # Check for identical rows
    print("\n [2] Check Duplicated")
    dupl = df.duplicated().sum()
    display(dupl)

    # Check column data types 
    print("\n [3] Data Type")
    columns_type = pd.DataFrame(
        {"Columns": df.columns, "Data Type": df.dtypes}
    )
    display(columns_type)

    # Statistical overview for numeric and categorical data
    print("\n [4] Describe")
    describe_df = df.describe(include= "all").T
    display(describe_df)
    
    # Identify missing values per column
    print("\n [5] Missing Data")
    null = pd.DataFrame({
        "Columns": df.columns, 
        "Missing Data": df.isnull().sum().values
    })
    display(null)

    # First 10 Rows
    print("\n [6] First Rows")
    display(df.head(10))

    # Last 10 Rows
    print("\n [7] Last Rows")
    display(df.tail(10))
    
    # Summary
    print("\n [8] Summary")
    summary = pd.DataFrame({
        "N_Rows": [df.shape[0]],
        "N_Columns": [len(df.columns)],
        "Memory": [df.memory_usage(deep= True).sum()]
    }) 
    display(summary)
```

# Key Data Quality Issues Summary – Olist Dataset

After initial exploration, here are the **most important observations** that require attention during the data cleaning phase:

## 1. Date / Timestamp Columns (Critical)
- **All date-related columns are stored as object (string)** instead of datetime  
  → Needs urgent conversion using `pd.to_datetime()`
- Affected columns (partial list):
  - orders: `order_purchase_timestamp`, `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`, `order_estimated_delivery_date`
  - order_items: `shipping_limit_date`
  - order_reviews: `review_creation_date`, `review_answer_timestamp`
- **Impact**: Cannot perform time-based analysis, calculations, or sorting correctly until converted

## 2. Missing Values (Missing Data)
| Table                     | Column(s) with missing values                  | Approx. missing count / %         | Priority    | Recommendation                              |
|---------------------------|------------------------------------------------|-----------------------------------|-------------|---------------------------------------------|
| orders                    | order_approved_at                              | 160                               | Medium      | Impute with purchase timestamp or drop rows |
| orders                    | order_delivered_carrier_date                   | ~1,783                            | High        | Important for delivery analysis             |
| orders                    | order_delivered_customer_date                  | ~2,965 (~3%)                      | High        | Critical for delivery time calculation      |
| products                  | product_category_name, name_lenght, description_lenght, photos_qty | 610 (~1.85%)     | Medium      | Impute category as 'unknown'                |
| products                  | weight_g, dimensions                           | 2                                 | Low         | Impute with median                          |
| order_reviews             | review_comment_title                           | ~87,656 (~88%)                    | High        | Almost always missing → drop or flag        |
| order_reviews             | review_comment_message                         | ~58,247 (~59%)                    | High        | High missing rate → consider text analysis only on available |

## 3. Duplicates / Uniqueness Issues
- **No critical duplicates** in primary keys across most tables (customer_id, order_id, product_id, seller_id, review_id are unique)
- **Major issue in geolocation table**:
  - `geolocation_zip_code_prefix` has **severe duplication** (~261,831 duplicates)
  - Cannot be used as a clean primary key or direct join key
  - **Solution already applied**: Created aggregated table `zipcode_locations` (unique zip prefix + averages)

## 4. Other Important Observations
- `customer_unique_id` is the **real customer identifier** (not `customer_id`)
  → Use it when analyzing customer behavior across multiple orders
- `product_category_name` in products has ~610 missing → must be handled before joining with translation table
- All text-based IDs (`object` dtype) are clean (no missing values)
- No negative values reported in price, freight_value, weight, etc. (still good to double-check)

## Summary of Priority Cleaning Actions
1. **Convert all timestamp columns** to proper datetime type
2. **Handle missing values** in delivery dates (orders) and review comments
3. **Use aggregated zipcode_locations** instead of raw geolocation table
4. **Impute or flag** missing product category and dimensions
5. **Consider dropping** or special handling for columns with very high missing rate (e.g., review_comment_title)

These are the **main pain points** that will affect analysis quality if not addressed.

## **Data Cleaning**

```python
# Handling Geolocation Duplicates: Removing exact duplicates to optimize the dataset
geolocation.drop_duplicates(inplace= True)
```

```python
# Data Type Correction: Converting string objects to datetime for time-series analysis
items.shipping_limit_date =             pd.to_datetime(items.shipping_limit_date)
o_reviews.review_creation_date =        pd.to_datetime(o_reviews.review_creation_date)
o_reviews.review_answer_timestamp =     pd.to_datetime(o_reviews.review_answer_timestamp)

# Bulk conversion for 'orders' table timestamps
date_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 
             'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors= 'coerce')
```

```python
# Handling Missing Values: Imputing missing comments and category info
o_reviews.review_comment_title.fillna('no_comment', inplace= True)
o_reviews.review_comment_message.fillna('no_comment', inplace= True)

# Imputing product categories with the most frequent value (Mode)
if products.product_category_name.isnull().any():
    products.product_category_name.fillna(products.product_category_name.mode()[0], inplace=True)

# Imputing numerical product specs with Median (robust to outliers)
num_cols = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']
for col in num_cols:
    products[col].fillna(products[col].median(), inplace=True)
```

```python
# Exporting cleaned datasets to CSV for the next stage (SQL/Power BI)
customers.to_csv('customers.csv', index= False)
geolocation.to_csv('geolocation.csv', index= False)
items.to_csv('order_items.csv', index= False)
payment.to_csv('order_payment.csv', index= False)
o_reviews.to_csv('order_review.csv', index= False)
orders.to_csv('orders.csv', index= False)
products.to_csv('products.csv', index= False)
product_category.to_csv('product_category.csv', index= False)
sellers.to_csv('sellers.csv', index= False)
```
